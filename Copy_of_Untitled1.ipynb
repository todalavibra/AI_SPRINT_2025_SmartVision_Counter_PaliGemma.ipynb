```python
# Load data
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load the breast cancer dataset
breast_cancer = load_breast_cancer()

# Convert to a pandas DataFrame
df = pd.DataFrame(data=breast_cancer.data, columns=breast_cancer.feature_names)
df['target'] = breast_cancer.target

# Preprocess data
from sklearn.preprocessing import StandardScaler

# Separate features (X) and target (y)
X = df.drop('target', axis=1)
y = df['target']

# Import and instantiate StandardScaler
scaler = StandardScaler()

# Fit and transform the feature data
X_scaled = scaler.fit_transform(X)

# Choose and train model
from sklearn.ensemble import RandomForestClassifier

# Instantiate the model
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_scaled, y)

# Evaluate model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions on the training data
y_pred = model.predict(X_scaled)

# Calculate and print the accuracy score
accuracy = accuracy_score(y, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Generate and print a classification report
print('Classification Report:')
print(classification_report(y, y_pred))

# Generate and print a confusion matrix
print('Confusion Matrix:')
print(confusion_matrix(y, y_pred))

# Make predictions (optional)
# Select a few rows as "new" data
new_data_df = df.sample(5, random_state=42)

# Separate features (X_new) and target (y_new)
X_new = new_data_df.drop('target', axis=1)
y_new = new_data_df['target']

# Scale the new data using the fitted scaler
X_new_scaled = scaler.transform(X_new)

# Make predictions on the scaled new data
predictions = model.predict(X_new_scaled)

# Print the predictions and compare with actual values
print("Predictions:", predictions)
print("Actual values:", y_new.values)
```

## Hyperparameter tuning

### Subtask:
Use techniques like GridSearchCV or RandomizedSearchCV to find the best hyperparameters for the model.


**Reasoning**:
Import GridSearchCV, define the hyperparameter grid, instantiate GridSearchCV with the model and grid, and fit it to the training data to find the best hyperparameters and their corresponding score.



```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters and best score
print("Best hyperparameters:", grid_search.best_params_)
print("Best cross-validation accuracy:", grid_search.best_score_)
```
